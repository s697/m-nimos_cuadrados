# 1. Dados los datos:

| \(x_i\) | 4.0 | 4.2 | 4.5 | 4.7 | 5.1 | 5.5 | 5.9 | 6.3 | 6.8 | 7.1 |
|----------|-----|-----|-----|-----|-----|-----|-----|-----|-----|-----|
| \(y_i\) | 102.56 | 130.11 | 113.18 | 142.05 | 167.53 | 195.14 | 224.87 | 256.73 | 299.50 | 326.72 |

a. Construya el polinomio por mínimos cuadrados de grado 1 y calcule el error.

b. Construya el polinomio por mínimos cuadrados de grado 2 y calcule el error.

c. Construya el polinomio por mínimos cuadrados de grado 3 y calcule el error.

d. Construya el polinomio por mínimos cuadrados de la forma \(be^{ax}\) y calcule el error.

e. Construya el polinomio por mínimos cuadrados de la forma \(bx^{a}\) y calcule el error.

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Datos
# -----------------------------

x = np.array([4.0,4.2,4.5,4.7,5.1,5.5,5.9,6.3,6.8,7.1])

y = np.array([102.56,
              130.11,
              113.18,
              142.05,
              167.53,
              195.14,
              224.87,
              256.73,
              299.50,
              326.72])

# -----------------------------
# Ajustes polinomiales
# -----------------------------

grado1 = np.polyfit(x,y,1)
grado2 = np.polyfit(x,y,2)
grado3 = np.polyfit(x,y,3)

p1 = np.poly1d(grado1)
p2 = np.poly1d(grado2)
p3 = np.poly1d(grado3)

# -----------------------------
# Ajuste exponencial
# y = b*e^(ax)
# ln(y)=ln(b)+ax
# -----------------------------

coef_exp = np.polyfit(x,np.log(y),1)

a_exp = coef_exp[0]
b_exp = np.exp(coef_exp[1])

# -----------------------------
# Ajuste potencial
# y=bx^a
# ln(y)=ln(b)+a ln(x)
# -----------------------------

coef_pot = np.polyfit(np.log(x),np.log(y),1)

a_pot = coef_pot[0]
b_pot = np.exp(coef_pot[1])

# -----------------------------
# Valores estimados
# -----------------------------

y1 = p1(x)
y2 = p2(x)
y3 = p3(x)

y_exp = b_exp*np.exp(a_exp*x)
y_pot = b_pot*x**a_pot

# -----------------------------
# Error cuadrático
# -----------------------------

def error(real,aprox):
    return np.sum((real-aprox)**2)

print("====================================")
print("POLINOMIO GRADO 1")
print(p1)
print("Error:",error(y,y1))

print()

print("POLINOMIO GRADO 2")
print(p2)
print("Error:",error(y,y2))

print()

print("POLINOMIO GRADO 3")
print(p3)
print("Error:",error(y,y3))

print()

print("MODELO EXPONENCIAL")
print(f"y={b_exp:.6f}e^({a_exp:.6f}x)")
print("Error:",error(y,y_exp))

print()

print("MODELO POTENCIAL")
print(f"y={b_pot:.6f}x^{a_pot:.6f}")
print("Error:",error(y,y_pot))

# -----------------------------
# Gráfica
# -----------------------------

xp = np.linspace(min(x),max(x),300)

plt.figure(figsize=(10,6))

plt.scatter(x,y,color="black",label="Datos")

plt.plot(xp,p1(xp),label="Grado 1")
plt.plot(xp,p2(xp),label="Grado 2")
plt.plot(xp,p3(xp),label="Grado 3")

plt.plot(xp,b_exp*np.exp(a_exp*xp),
         label="Exponencial")

plt.plot(xp,b_pot*xp**a_pot,
         label="Potencial")

plt.title("Ejercicio 1 - Ajustes por mínimos cuadrados")
plt.xlabel("x")
plt.ylabel("y")

plt.grid(True)
plt.legend()

plt.show()
```

# 2. Repita el ejercicio 5 para los siguientes datos.

| \(x_i\) | 0.2 | 0.3 | 0.6 | 0.9 | 1.1 | 1.3 | 1.4 | 1.6 |
|----------|-----|-----|-----|-----|-----|-----|-----|-----|
| \(y_i\) | 0.050446 | 0.098426 | 0.33277 | 0.72660 | 1.0972 | 1.5697 | 1.8487 | 2.5015 |

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Datos
# -----------------------------

x = np.array([0.2,0.3,0.6,0.9,1.1,1.3,1.4,1.6])

y = np.array([
    0.050446,
    0.098426,
    0.332770,
    0.726600,
    1.097200,
    1.569700,
    1.848700,
    2.501500
])

# -----------------------------
# Ajustes polinomiales
# -----------------------------

coef1 = np.polyfit(x, y, 1)
coef2 = np.polyfit(x, y, 2)
coef3 = np.polyfit(x, y, 3)

p1 = np.poly1d(coef1)
p2 = np.poly1d(coef2)
p3 = np.poly1d(coef3)

# -----------------------------
# Ajuste exponencial
# y = b e^(ax)
# -----------------------------

coef_exp = np.polyfit(x, np.log(y), 1)

a_exp = coef_exp[0]
b_exp = np.exp(coef_exp[1])

# -----------------------------
# Ajuste potencial
# y = b x^a
# -----------------------------

coef_pot = np.polyfit(np.log(x), np.log(y), 1)

a_pot = coef_pot[0]
b_pot = np.exp(coef_pot[1])

# -----------------------------
# Valores estimados
# -----------------------------

y1 = p1(x)
y2 = p2(x)
y3 = p3(x)

y_exp = b_exp * np.exp(a_exp * x)
y_pot = b_pot * x**a_pot

# -----------------------------
# Error cuadrático
# -----------------------------

def error(real, estimado):
    return np.sum((real-estimado)**2)

print("="*50)

print("POLINOMIO GRADO 1")
print(p1)
print("Error:", error(y, y1))

print()

print("POLINOMIO GRADO 2")
print(p2)
print("Error:", error(y, y2))

print()

print("POLINOMIO GRADO 3")
print(p3)
print("Error:", error(y, y3))

print()

print("MODELO EXPONENCIAL")
print(f"y = {b_exp:.6f} e^({a_exp:.6f}x)")
print("Error:", error(y, y_exp))

print()

print("MODELO POTENCIAL")
print(f"y = {b_pot:.6f} x^{a_pot:.6f}")
print("Error:", error(y, y_pot))

# -----------------------------
# Gráfica
# -----------------------------

xp = np.linspace(min(x), max(x), 300)

plt.figure(figsize=(10,6))

plt.scatter(x, y, color="black", s=50, label="Datos")

plt.plot(xp, p1(xp), label="Grado 1")
plt.plot(xp, p2(xp), label="Grado 2")
plt.plot(xp, p3(xp), label="Grado 3")
plt.plot(xp, b_exp*np.exp(a_exp*xp), label="Exponencial")
plt.plot(xp, b_pot*xp**a_pot, label="Potencial")

plt.title("Ejercicio 2 - Ajuste por mínimos cuadrados")
plt.xlabel("x")
plt.ylabel("y")

plt.grid(True)
plt.legend()

plt.show()
```

# 3.

La siguiente tabla muestra los promedios de puntos del colegio de 20 especialistas en matemáticas y ciencias computacionales, junto con las calificaciones que recibieron estos estudiantes en la parte de matemáticas de la prueba ACT (Programa de Pruebas de Colegios Americanos) mientras estaban en secundaria. Grafique estos datos y encuentre la ecuación de la recta por mínimos cuadrados para estos datos.

| Puntuación ACT | Promedio de puntos |
|----------------|-------------------|
| 28 | 3.84 |
| 25 | 3.21 |
| 28 | 3.23 |
| 27 | 3.63 |
| 28 | 3.75 |
| 33 | 3.20 |
| 28 | 3.41 |
| 29 | 3.38 |
| 23 | 3.53 |
| 27 | 2.03 |
| 29 | 3.75 |
| 28 | 3.65 |
| 27 | 3.87 |
| 29 | 3.75 |
| 21 | 1.66 |
| 28 | 3.12 |
| 28 | 2.96 |
| 26 | 2.92 |
| 30 | 3.10 |

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------------
# Datos
# -----------------------------------

x = np.array([
28,25,28,27,28,
33,28,29,23,27,
29,28,27,29,21,
28,28,26,30,24
])

y = np.array([
3.84,3.21,3.23,3.63,3.75,
3.20,3.41,3.38,3.53,2.03,
3.75,3.65,3.87,3.75,1.66,
3.12,2.96,2.92,3.10,2.81
])

# -----------------------------------
# Recta por mínimos cuadrados
# -----------------------------------

coef = np.polyfit(x, y, 1)

m = coef[0]
b = coef[1]

recta = np.poly1d(coef)

# -----------------------------------
# Valores estimados
# -----------------------------------

y_est = recta(x)

# -----------------------------------
# Error cuadrático
# -----------------------------------

error = np.sum((y - y_est)**2)

print("="*40)
print("RECTA POR MINIMOS CUADRADOS")
print("="*40)

print(f"Pendiente (m) = {m:.6f}")
print(f"Intercepto (b) = {b:.6f}")

print()
print(f"Ecuación:")

print(f"y = {m:.6f}x + {b:.6f}")

print()
print("Error cuadrático:", error)

# -----------------------------------
# Gráfica
# -----------------------------------

xp = np.linspace(min(x), max(x), 300)

plt.figure(figsize=(9,6))

plt.scatter(
    x,
    y,
    color="black",
    s=60,
    label="Datos"
)

plt.plot(
    xp,
    recta(xp),
    linewidth=2,
    label="Recta de mínimos cuadrados"
)

plt.title("Ejercicio 3")
plt.xlabel("Puntuación ACT")
plt.ylabel("Promedio")

plt.grid(True)
plt.legend()

plt.show()
```

# 4.

El siguiente conjunto de datos, presentado al Subcomité Antimonopolio del Senado, muestra las características comparativas de supervivencia durante un choque de automóviles de diferentes clases. Encuentre la recta por mínimos cuadrados que aproxima estos datos (la tabla muestra el porcentaje de vehículos que participaron en un accidente en los que la lesión más grave fue fatal o seria).

| Tipo | Peso promedio | Porcentaje de presentación |
|--------|--------------|---------------------------|
| 1. Regular lujoso doméstico | 4800 lb | 3.1 |
| 2. Regular intermediario doméstico | 3700 lb | 4.0 |
| 3. Regular económico doméstico | 3400 lb | 5.2 |
| 4. Compacto doméstico | 2800 lb | 6.4 |
| 5. Compacto extranjero | 1900 lb | 9.6 |

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------------
# Datos
# -----------------------------------

x = np.array([4800,3700,3400,2800,1900])

y = np.array([3.1,4.0,5.2,6.4,9.6])

# -----------------------------------
# Ajuste lineal
# -----------------------------------

coef = np.polyfit(x, y, 1)

m = coef[0]
b = coef[1]

recta = np.poly1d(coef)

# -----------------------------------
# Valores estimados
# -----------------------------------

y_est = recta(x)

# -----------------------------------
# Error cuadrático
# -----------------------------------

error = np.sum((y - y_est)**2)

print("="*45)
print("RECTA POR MÍNIMOS CUADRADOS")
print("="*45)

print(f"Pendiente (m): {m:.6f}")
print(f"Intercepto (b): {b:.6f}")

print()
print(f"Ecuación:")

print(f"y = {m:.6f}x + {b:.6f}")

print()
print(f"Error cuadrático: {error:.6f}")

# -----------------------------------
# Gráfica
# -----------------------------------

xp = np.linspace(min(x), max(x), 300)

plt.figure(figsize=(9,6))

plt.scatter(
    x,
    y,
    color="red",
    s=70,
    label="Datos"
)

plt.plot(
    xp,
    recta(xp),
    linewidth=2,
    label="Recta de mínimos cuadrados"
)

plt.title("Ejercicio 4")
plt.xlabel("Peso promedio (lb)")
plt.ylabel("Porcentaje de accidentes (%)")

plt.grid(True)
plt.legend()

plt.show()
```
